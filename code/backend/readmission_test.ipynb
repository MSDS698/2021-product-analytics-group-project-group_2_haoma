```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_excel('readmission_1.xlsx')
df_sample = df.sample(frac=0.7)
```

```python
df_sample
```

```python
fig = px.histogram(df_sample, x="Date", color="Gender", color_discrete_sequence=['#DB4437', '#4285F4'], opacity=0.75)
fig.update_layout(
    title_text='Number of Readmission', # title of plot
    xaxis_title_text='Date', # xaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)
fig.show()
```

```python
fig = px.histogram(df_sample, y="Service")
fig.show()
```
